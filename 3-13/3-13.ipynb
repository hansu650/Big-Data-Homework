```python
from pathlib import Path
data_dir = Path('mydata')
data_dir.mkdir(exist_ok=True)  # 目录不存在就创建，已存在则不报错
csv_path = data_dir / 'foo.csv'
df1.to_csv(csv_path)  # 导出为 CSV 文件
```

```python
df = pd.read_csv(csv_path)  # 从 CSV 文件读回 DataFrame
print(df[:10])
df
```

```python
hdf_path = data_dir / 'foo.h5'
try:
    df.to_hdf(hdf_path, key='df')  # 保存为 HDF5，key 类似文件里的数据表名
    print(f'Saved HDF5 file: {hdf_path}')
except ImportError as exc:
    print(f'Skipping HDF5 write: {exc}')
```

```python
if hdf_path.exists():
    df = pd.read_hdf(hdf_path, key='df')  # 从 HDF5 中读取 key='df' 的数据
    print(df[:10])
else:
    print('Skipping HDF5 read because pytables is not installed.')
```

```python
excel_path = data_dir / 'foo.xlsx'
try:
    df.to_excel(excel_path, sheet_name='Sheet01')  # 导出到 Excel 的指定工作表
    print(f'Saved Excel file: {excel_path}')
except ModuleNotFoundError as exc:
    print(f'Skipping Excel write: {exc}')
```

```python
if excel_path.exists():
    print(pd.read_excel(excel_path, 'Sheet01', index_col=None, na_values=['NA'])[:10])  # 读取 Excel 并把 'NA' 识别成缺失值
else:
    print('Skipping Excel read because openpyxl is not installed.')
```

```python
rng = pd.date_range('1/1/2012', periods=100, freq='s')  # 每秒生成一个时间点
print(rng)
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
print(ts.head(5))
ts_sample = ts.resample('5Min')  # 按 5 分钟重采样，先得到重采样对象
print(ts_sample)
ts_sample.sum()
```

```python
rng = pd.date_range('3/6/2012 00:00', periods=5, freq='D')
ts = pd.Series(np.random.randn(len(rng)), rng)
print(ts)
ts_utc = ts.tz_localize('UTC')  # 给原始时间序列附加 UTC 时区信息
ts_utc
```

```python
rng = pd.date_range('1/1/2012', periods=5, freq='s')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
print(ts)
ps = ts.to_period()  # 把时间戳索引转换成 PeriodIndex
print(ps)
ps.to_timestamp()  # 再从 PeriodIndex 转回时间戳索引
```

```python
df = pd.DataFrame({
    "revenue": [57, 68, 63, 71, 72, 90, 80, 62, 59, 51, 47, 52],
    "advertising": [2.1, 1.9, 2.7, 3.0, 3.6, 3.2, 2.7, 2.4, 1.8, 1.6, 1.3, 1.9],
    "month": range(12)
})
ax = df.plot.bar("month", "revenue", color="green")  # 先画营业额柱状图
df.plot.line("month", "advertising", secondary_y=True, ax=ax)  # secondary_y=True 表示广告费走右侧 y 轴
ax.set_xlim((-1, 12))
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts.plot()  # plot 会按时间索引自动绘制折线图
```
